Set 1  Challenge 1 : Convert hex to base64

```python
from base64 import b16decode, b64encode

data_hex = b"49276d206b696c6c696e6720796f757220627261696e206c696b65206120706f69736f6e6f7573206d757368726f6f6d"
data_raw = b16decode(data_hex, casefold=True)
data_b64 = b64encode(data_raw)

print(f"{data_hex=}")
print(f"{data_raw=}")
print(f"{data_b64=}")

if data_b64 == b"SSdtIGtpbGxpbmcgeW91ciBicmFpbiBsaWtlIGEgcG9pc29ub3VzIG11c2hyb29t":
    print("It worked!!!")
else:
    print("Conversion failed")
```

Set 1 Challenge 2 : Fixed XOR

```python
def xor(a: str,b: str) -> str:
    if len(a) != len(b):
        raise ValueError("Input strings not equal length")

    return hex(int(a, 16) ^ int(b, 16))[2:] 

def main():
    hex1 = '1c0111001f010100061a024b53535009181c'
    hex2 = '686974207468652062756c6c277320657965'
    ans = '746865206b696420646f6e277420706c6179'

    result = xor(hex1, hex2)

    print(f"correct? {ans == result}, {result}")

if __name__ == '__main__':
    main()
```
